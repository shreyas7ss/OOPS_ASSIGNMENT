```python
from transformers import AutoTokenizer
from datasets import load_dataset

# Load dataset
dataset = load_dataset("cnn_dailymail", "3.0.0")

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Define max lengths
MAX_ARTICLE_LEN = 512
MAX_SUMMARY_LEN = 128
```

```python
def tokenize_function(example):
    # Tokenize article (input to model)
    inputs = tokenizer(
        example["article"],         # The original article text
        truncation=True,            # Truncate if it exceeds max_length
        padding="max_length",       # Pad shorter sequences to max_length
        max_length=MAX_ARTICLE_LEN  # Fixed max length = 512
    )

    # Tokenize summary (output/label)
    labels = tokenizer(
        example["highlights"],       # The summary text
        truncation=True,             # Truncate if too long
        padding="max_length",        # Pad if too short
        max_length=MAX_SUMMARY_LEN   # Fixed max length = 128
    )

    # Return tokenized data
    return {
        "input_ids": inputs["input_ids"],  # Token IDs for article
        "attention_mask": inputs["attention_mask"],  # Tells model where to focus
        "labels": labels["input_ids"]  # Token IDs for the summary (target)
    }

tokenized_train = dataset["train"].map(tokenize_function, batched=True)
tokenized_validation = dataset["validation"].map(tokenize_function, batched=True)
tokenized_test = dataset["test"].map(tokenize_function, batched=True)

tokenized_datasets = {
    "train": tokenized_train,
    "validation": tokenized_validation,
    "test": tokenized_test
}


```

```python
import tensorflow as tf

# Convert to TensorFlow Dataset
train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(
        input_ids=tokenized_datasets["train"]["input_ids"],
        attention_mask=tokenized_datasets["train"]["attention_mask"]
    ),
    tokenized_datasets["train"]["labels"]
))

val_dataset = tf.data.Dataset.from_tensor_slices((
    dict(
        input_ids=tokenized_datasets["validation"]["input_ids"],
        attention_mask=tokenized_datasets["validation"]["attention_mask"]
    ),
    tokenized_datasets["validation"]["labels"]
))

# Shuffle & batch the dataset
BATCH_SIZE = 8

train_dataset = train_dataset.shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_dataset = val_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# Print sample
for batch in train_dataset.take(1):
    print(batch)
```

```python
import tensorflow as tf
import numpy as np

# Positional Encoding
class PositionalEncoding(tf.keras.layers.Layer):
    def __init__(self, d_model, max_len=5000):
        super().__init__()
        self.d_model = d_model
        self.pe = self.positional_encoding(max_len, d_model)

    def positional_encoding(self, max_len, d_model):
        pos = np.arange(max_len)[:, np.newaxis]
        div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))
        pe = np.zeros((max_len, d_model))
        pe[:, 0::2] = np.sin(pos * div_term)
        pe[:, 1::2] = np.cos(pos * div_term)
        return tf.convert_to_tensor(pe[np.newaxis, ...], dtype=tf.float32)

    def call(self, x):
        return x + self.pe[:, :tf.shape(x)[1], :]

# Multi-Head Attention
class MultiHeadAttention(tf.keras.layers.Layer):
    def __init__(self, d_model, num_heads):
        super().__init__()
        self.num_heads = num_heads
        self.d_model = d_model
        self.depth = d_model // num_heads

        self.wq = tf.keras.layers.Dense(d_model)
        self.wk = tf.keras.layers.Dense(d_model)
        self.wv = tf.keras.layers.Dense(d_model)
        self.dense = tf.keras.layers.Dense(d_model)

    def split_heads(self, x, batch_size):
        x = tf.reshape(x, (batch_size, -1, self.num_heads, self.depth))
        return tf.transpose(x, perm=[0, 2, 1, 3])

    def call(self, v, k, q, mask):
        batch_size = tf.shape(q)[0]
        q = self.split_heads(self.wq(q), batch_size)
        k = self.split_heads(self.wk(k), batch_size)
        v = self.split_heads(self.wv(v), batch_size)

        scores = tf.matmul(q, k, transpose_b=True) / tf.math.sqrt(tf.cast(self.depth, tf.float32))
        if mask is not None:
            scores += (mask * -1e9)
        attention_weights = tf.nn.softmax(scores, axis=-1)
        output = tf.matmul(attention_weights, v)
        output = tf.transpose(output, perm=[0, 2, 1, 3])
        output = tf.reshape(output, (batch_size, -1, self.d_model))
        return self.dense(output)

# Transformer Encoder Layer
class EncoderLayer(tf.keras.layers.Layer):
    def __init__(self, d_model, num_heads, dff, dropout_rate=0.1):
        super().__init__()
        self.mha = MultiHeadAttention(d_model, num_heads)
        self.ffn = tf.keras.Sequential([
            tf.keras.layers.Dense(dff, activation='relu'),
            tf.keras.layers.Dense(d_model)
        ])
        self.layernorm1 = tf.keras.layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = tf.keras.layers.LayerNormalization(epsilon=1e-6)
        self.dropout1 = tf.keras.layers.Dropout(dropout_rate)
        self.dropout2 = tf.keras.layers.Dropout(dropout_rate)

    def call(self, x, mask):
        attn_output = self.mha(x, x, x, mask)
        attn_output = self.dropout1(attn_output)
        out1 = self.layernorm1(x + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output)
        return self.layernorm2(out1 + ffn_output)

# Transformer Encoder
class Encoder(tf.keras.layers.Layer):
    def __init__(self, num_layers, d_model, num_heads, dff, vocab_size, max_len, dropout_rate=0.1):
        super().__init__()
        self.embedding = tf.keras.layers.Embedding(vocab_size, d_model)
        self.pos_encoding = PositionalEncoding(d_model, max_len)
        self.enc_layers = [EncoderLayer(d_model, num_heads, dff, dropout_rate) for _ in range(num_layers)]
        self.dropout = tf.keras.layers.Dropout(dropout_rate)

    def call(self, x, mask):
        x = self.embedding(x)
        x = self.pos_encoding(x)
        x = self.dropout(x)
        for layer in self.enc_layers:
            x = layer(x, mask)
        return x

# Define Hyperparameters
NUM_LAYERS = 6
D_MODEL = 128
NUM_HEADS = 8
DFF = 768
VOCAB_SIZE = 30000
MAX_LEN = 512
DROPOUT_RATE = 0.1

# Build the Transformer Encoder Model
encoder = Encoder(NUM_LAYERS, D_MODEL, NUM_HEADS, DFF, VOCAB_SIZE, MAX_LEN, DROPOUT_RATE)

# Example Input
sample_input = tf.random.uniform((1, MAX_LEN), minval=0, maxval=VOCAB_SIZE, dtype=tf.int32)
mask = tf.cast(tf.math.equal(sample_input, 0), tf.float32)[:, tf.newaxis, tf.newaxis, :]
output = encoder(sample_input, mask)
print(output.shape)  # Expected: (1, MAX_LEN, D_MODEL)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, LayerNormalization, Dropout
from tensorflow.keras.models import Model

class DecoderLayer(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, dropout_rate=0.1):
        super(DecoderLayer, self).__init__()
        self.self_attention = tf.keras.layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.norm1 = LayerNormalization(epsilon=1e-6)
        self.dropout1 = Dropout(dropout_rate)

        self.cross_attention = tf.keras.layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.norm2 = LayerNormalization(epsilon=1e-6)
        self.dropout2 = Dropout(dropout_rate)

        self.ffn = tf.keras.Sequential([
            Dense(ff_dim, activation="relu"),
            Dense(embed_dim)
        ])
        self.norm3 = LayerNormalization(epsilon=1e-6)
        self.dropout3 = Dropout(dropout_rate)

    def call(self, x, enc_output, training, look_ahead_mask, padding_mask):
        # Masked Self-Attention (Decoder looks only at past words)
        attn1 = self.self_attention(x, x, x, attention_mask=look_ahead_mask)
        attn1 = self.dropout1(attn1, training=training)
        out1 = self.norm1(x + attn1)

        # Cross-Attention (Decoder attends to Encoder output)
        attn2 = self.cross_attention(out1, enc_output, enc_output, attention_mask=padding_mask)
        attn2 = self.dropout2(attn2, training=training)
        out2 = self.norm2(out1 + attn2)

        # Feed Forward Network
        ffn_output = self.ffn(out2)
        ffn_output = self.dropout3(ffn_output, training=training)
        out3 = self.norm3(out2 + ffn_output)

        return out3
```

```python
import tensorflow as tf

loss_object = tf.keras.losses.SparseCategoricalCrossentropy(
    from_logits=True,  # Since our model outputs raw logits
    reduction='none'   # We handle masking manually
)

def loss_function(real, pred):
    """
    Compute the loss, ignoring padding tokens.
    """
    mask = tf.math.logical_not(tf.math.equal(real, 0))  # Create mask where real ≠ 0 (non-padding)
    loss = loss_object(real, pred)

    mask = tf.cast(mask, dtype=loss.dtype)  # Convert mask to correct dtype
    loss *= mask  # Apply mask to loss

    return tf.reduce_mean(loss)  # Average loss across batch
```

```python
train_accuracy = tf.keras.metrics.SparseCategoricalAccuracy(name="train_accuracy")

def accuracy_function(real, pred):
    """
    Compute accuracy, ignoring padding tokens.
    """
    predictions = tf.argmax(pred, axis=-1)  # Get predicted token IDs
    mask = tf.math.logical_not(tf.math.equal(real, 0))  # Ignore padding
    mask = tf.cast(mask, dtype=tf.float32)

    accuracy = tf.cast(tf.equal(real, predictions), dtype=tf.float32)
    accuracy *= mask  # Apply mask

    return tf.reduce_sum(accuracy) / tf.reduce_sum(mask)  # Only consider non-padding tokens
```
